Each single image can be represented as an array. As an image get's more and more pixels, we will have finer and finer resolution. Values between 0 and 1 represents how dark a pixel should be. The place where it is completely dark is represented as 1 and the place where it is completely white, it's represented as 0. 

Often default images have values between 0 and 255. 255 represents the highest or brightest value and zero is the darkest. The range 0 to 255 has to do with how computers store 8-bit numbers. For converting all values ranging from 0 to 255 to the values ranging between 0 and 1, we need to divide all the values by 255. This is how we represent grayscale images can be represented as arrays.

Color images can be represented as a combination of Red, Green and Blue. Additive color mixing allows us to represent a wide variety of colors by simply combining different amounts of Red, Green , Blue or R, G, B. RGB allows us to produce a range of colors. 

Color Triangle : Only colors within this triangle can be reproduced by mixing the primary colors i.e. red, green, blue. Colors outside the traingle are shown as grey. 

Each color channel will have intensity values. The shape of the color array then has 3 dimensions namely height, width and color channels. This means that when we read in an image, it will have 3 dimensions and those dimensions are the ones mentioned above. Example : If we read in an image and we find out after checking that it's shape is (1280,720,3), then this means that :-

1280 is the pixel width, 

720 is the pixel height, and,

3 is the color channels.

The computer does not know which channel represents which color. The user needs to dictate which channel is for which color. Each channel alone is essentially the same as a grayscale image. 

Numpy by itself cannot directly open a jpeg or png file. We need to use it in conjuction with PIL(Python Imaging Library). 

```python
# %matplotlib inline is written to ensure that our plots which are plotted using matplotlib remain in this notebook itself
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from PIL import Image # This is an image function that allows us to open up images and then we are going to be able to transform this image into an array that numpy can understand
```

```python
pic = Image.open('lr.png')
```

```python
type(pic) # Since it is a png file, numpy cannot deal with it. So we need to convert it into an array. 
```

```python
pic
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
pic_red.shape
```

```python
# Red channel values 0- no red pure black, 255- full pure red
plt.imshow(pic_red[:,:,0],cmap = 'gray')
```

```python
# Green channel values
plt.imshow(pic_red[:,:,1],cmap = 'gray')
```

```python
# Blue channel values
plt.imshow(pic_red[:,:,2],cmap = 'gray')
```

```python
pic_red[:,:,1] = 0 # Making all green channel values as 0
```

```python
plt.imshow(pic_red) # It looks purplish since the only 2 colors that are left are red and blue since we made green channel as 0
```

```python
pic_red[:,:,2] = 0 # Now making all blue channel values as 0 i.e. contribution of blue color also becomes 0 in the image
```

```python
plt.imshow(pic_red) # Since we made the contribution of blue and green color as 0, this means that now only red color's contribution is there.
```
