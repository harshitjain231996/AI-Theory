```python
# Data Preprocessing will be done first
import numpy as np # Library that contains mathematical tools and will be used to do any mathematics in code
import pandas as pd # Library used for importing datasets and managing them. It's really intuitive as well
import matplotlib.pyplot as plt # pyplot is a sub-library of matplotlib. Used for plotting nice charts.
dataset = pd.read_csv('C:/Users/Admin/Desktop/Python Notebooks/Machine Learning Theory/Machine Learning A-Z/Part 2 - Regression/Section 4 - Simple Linear Regression/Salary_Data.csv')
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:,1].values
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)
```

```python
# We do not need to do feature scaling in linear regression
# Fitting simple linear regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # Object of linear regression class that will be fit to the training set. The function with empty parameters is used so that it will create a self function of itself
regressor.fit(X_train,Y_train) # Fitting the linear regression method to training set 
```

```python
#Predicting the test set results. We have already fitted the regressor on the training set and the machine has learnt about the linear regression model with the training data. That's why we are testing it's learning performance via providing the test data.
Y_pred = regressor.predict(X_test) # Vector of predictions of the dependent variable
Y_pred
```

```python
#Visualizing the training set results
plt.scatter(X_train,Y_train,color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing the test set results
plt.scatter(X_test,Y_test, color = 'red')
plt.plot(X_train,regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
