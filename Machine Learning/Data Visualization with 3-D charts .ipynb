# Data Visualization with 3d charts

## Minimize $$f(x,y) = \frac{1}{3^{(-x^2-y^2)}+1}$$

Another method

Minimize $$f(x,y) = \frac{1}{r+1}$$, where $r$ is $3^{-x^2-y^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm # cm stands for color map
```

```python
def f(x,y) :
    r = 3**(-x**2 - y**2)
    return 1/(r+1)
```

```python
#Making data for both x and y
x_4 = np.linspace(start = -2, stop = 2, num = 200) # linspace function returns an ndarray
y_4 = np.linspace(start = -2 ,stop = 2,num = 200)
print("shape of x array: ",x_4.shape)
x_4, y_4 = np.meshgrid(x_4,y_4)
print("Array after meshgrid:",x_4.shape)
```

```python
# generating 3d plot
fig = plt.figure(figsize=[16,12]) # figure is conceptually a matplotlib object that allows us to manipulate our final chart in some ways like changing the signs or adding axis to the chart. We can think of figure as top level container that contains other elements in the chart.
ax=fig.gca(projection='3d') # gca stands for get current axis
ax.plot_surface(x_4,y_4,f(x_4,y_4),cmap=cm.winter,alpha=0.4) # for different color maps, we can look at the official documentation to change and fit another cm
ax.set_xlabel('X',fontsize=20)
ax.set_ylabel('Y',fontsize=20)
ax.set_zlabel('f(x,y)-Cost',fontsize=20)
plt.show()
```
