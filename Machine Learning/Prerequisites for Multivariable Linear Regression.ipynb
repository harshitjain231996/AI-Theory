# Using Histogram as a data visualization tool

Histogram is being used as a data visualization tool since it takes care of 2 things:
1. Helps us see the distribution of the data
2. To spot the outliers

Histogram shows a frequency of a particular value. The taller an individual bar, the more are the number of occurences in the dataset. If the histogram is like a bell-curve, then we can see that a lot of data is around the mean or center of the distribution, and very few observations are at the edges.

```python
plt.figure(figsize=(10,6))
plt.hist(data['PRICE'],bins=50,ec='black',color='#2196f3') # ec- edge color
plt.xlabel('Price in 1000\'s')
plt.ylabel('Number of houses')
plt.show()
```

# Visualizing data with heat maps

Multicollinearity - When 2 or more predictors in a regression are highly related to one another. Each of these do not provide unique and independent information to the regression. 

High correlation does not necessarily mean we have problem of multicollinearity. High correlation between 2 features may or may not be a problem.

We should be looking at correlations between our features, we should be investigating if our features are highly correlated and if we find a high correlation between our features, we should investigate why that is.

High correlation can be an early warning sign that there is some sort of problem .

```python
mask = np.zeros_like(data.corr())
mask
```

```python
traingle_indices = np.triu_indices_from(mask) # Will return the top traingle indices of the array
mask[traingle_indices] = True
mask
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(data.corr(),mask=mask,annot=True,annot_kws={"size":14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

If we look at the heat map above, we see that the highest correlation is between TAX and RAD. Heatmap assumes some kind of assumption between the kind of data that it's running on. The correlation calculation is only valid for continuous variables and it's not valid for discrete variables. Since RAD is a discrete variable and not a continuous variable, so the correlation calculation is not valid for RAD feature. We can only use statistical tools for the kind of data we are  working with. Hence, it is highly important to know about the type of data we are dealing with. 

It is highly important to use plots in conjunction with descriptive statistics to spot patterns and outliers.

If a feature is not adding any explanatory value to the regression, it should better be excluded and we should run the regression without it because by excluding features, we might end up with a simpler model. 

Correlation does not imply causation. Just because 2 things move together, it does not mean one thing causes another.

Correlations only check for linear relationships.

Outliers and non-linear relationships often only become apparent after visualizing the data.

# Styling scatter plots

```python
nox_dis_corr = round(data['NOX'].corr(data['DIS']),4)
plt.figure(figsize=(10,6))
plt.title(f'DIS v/s NOX- The correlation = {nox_dis_corr}',fontsize=20)
plt.xlabel('DIS- Distance from Employment centre',fontsize=14)
plt.ylabel('NOX- Nitrix Oxide Concentration',fontsize=14)
plt.scatter(data['DIS'],data['NOX'],alpha=0.6,s=80,color='indigo')
```

```python
sns.set() # Resets the style since jupyter notebook remembers the style
sns.set_context('talk')
sns.set_style('whitegrid') # We can choose the style and modify the graph accordingly
sns.jointplot(x=data['DIS'],y=data['NOX'],height=7,color='indigo',joint_kws={'alpha': 0.5})
plt.show()
```

```python
sns.set() # Resets the style since jupyter notebook remembers the style
sns.set_context('talk')
sns.set_style('whitegrid') # We can choose the style and modify the graph accordingly
sns.jointplot(x=data['DIS'],y=data['NOX'],height=7,color='blue',kind='hex')
plt.show()
```

```python
sns.set() # Resets the style since jupyter notebook remembers the style
sns.set_context('talk')
sns.set_style('whitegrid') # We can choose the style and modify the graph accordingly
sns.jointplot(x=data['TAX'],y=data['RAD'],height=7,color='darkred',joint_kws={'alpha': 0.5})
plt.show()
```
