# Bag Of Words

Treats each document as an unordered collection or bag of words. The document is the unit of text we want to analyze.

We are turning each document into a vector of numbers representing how many times each word occurs in a document. A set of words is called a corpus. We will first collect all the unique words from the corpus to form the vocabulary. We will arrange the words in some order and they will take the position of the column. The row will be representing each document. After this, we will count the number of occurences of each word in each document and enter the value in the respective column. This will lead to the formation of document term matrix which illustrates the relationship between words in columns and document in rows. Each element in the matrix can be interpreted as a term frequency i.e. how frequently does that term occur in this document. For calculating the similarity between 2 documents, we will use cosine similarity.

cos(theta) = [(dot product of the two documents)]/[(modulus of 1st document)*(modulus of 2nd document)]

The cosine similarity lies between -1 and +1. If the cosine similarity between 2 documents is +1, it means that the documents are 100% similar. If the cosine similarity between the documents is -1, it means that the documents are completely dissimilar to each other.

# One-Hot Encoding

A way of representing words. Used when we talk about categorical variables. Works only in some situations but breaks down when we have a large vocabulary to deal with because the size of our word representation grows with the number of words.

# Word Embedding

We need to control the size of our word representation by limiting it to a fixed size vector. In other words, we need to find an embedding for each word in the vector space and we want it to exhibit some desired property. For instance - If words have similar meanings, then they should be closer. Also if the two pair of words have similar difference in their meanings, they should be approximately equally separated in the embedded space. This kind of representation can be used for finding synonyms,classifying words whether they are positive, negative or neutral, identifying the concepts around which the words are clustered and so on.

# Word2Vec

One of the most popular technique of word embedding used in practice. As the name indicates, it transforms words to vectors. The model that is able to predict the given word given neighbouring words or vice versa, is likely to capture the contextual meanings of the words very well.
Continuous Bag of words -  We are given neighbouring words
Continuous Skip Gram - We are given the middle word

In the skip gram model, we take any word, convert it into a one-hot encoded vector, and feed it into a neural network or some probabilistic model that is designed to predict a few surrounding words i.e. it's context. Using the suitable loss functions, optimizie the weights of the model and repeat this till it learns to predict context words as best as it can. Take a intermediate stage like hidden layer in neural network and the outputs for that layer in the given word become the corresponding word vector. The Continuous bag of words uses the same approach.

This yields a robust and distributed representation. Word2Vec gives a vector size which is independent of the vocabulary unlike in Bag of words where the size will change according to the words present. Once Word2vec is trained, then it can be stored in a lookup table. The result of word2vec can be used directly in recurrent neural networks, so it is deep learning ready.

# GloVe

GLobal Vectors for word representation(GloVe) is a type of word embedding. It tries directly to optimize the vector representation of each word using cooccurence vector.

First we compute the probability of word j appears in the context of word i is computed. We are then initializing two vectors, one for the context and the other one for the target. Then for a given word, we want the w of i and w of j's dot product to be equal to the value of their cooccurence matrix. The result should capture the similarities and differences between individual words.




# Cooccurence Probabilities




The cooccurence probability matrix is huge and at the same time cooccurence probability values are typically low, therefore it makes sense to work with the log of these values.

When a large collection of sentences is used to learn in embedding,words with common context tend to get pulled closer and closer together. Inorder to differenciate between them, we add another dimension. The more dimensions we can capture in our word vector, the more expressive that representation will be. Best way is to use pretrained models so that we get to know the optimal amount of dimensions which are required to make the representation expressive.

# T-SNE

Word embeddings need to have high dimensionality in order to capture sufficient variance in natural language and this makes them hard to visualize. T-distributed Stochastic Neighbor Embedding is a dimensionality reduction technique that can map high dimensional vectors to a lower dimensional space. Very similar to PCA. When performing the transformation, it tries to maintain relative distances between objects so that similar ones stay closer together while dissimilar object stay further apart. It preserves the relationship that have been learnt by embedding model. It works on images as well. So it will make a cluster of similar images.
