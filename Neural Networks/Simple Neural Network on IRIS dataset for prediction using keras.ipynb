```python
import numpy as np
```

```python
import keras 
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
X = iris.data # X is the features
```

```python
y = iris.target # y is the target value
```

```python
y
```

Because of the way keras and neural network work in general, we need to convert the above to categorical labelling system essentially one-hot encoding. Example : 
class 0 ---> [1,0,0]
class 1 ---> [0,1,0]
class 2 ---> [0,0,1]

This is called one-hot encoding. We will convert the above data by using keras.

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#MinMaxScaler converts all the values to fall in the range of [0,1]

#Example of how MinMaxScaler works:
np.array([5,10,15,20])/20
# We get all the values lying between 0 and 1 and that's exactly how MinMaxScaler performs the conversion.
```

```python
scaler_object = MinMaxScaler()
```

```python
scaler_object.fit(X_train)
```

```python
scaled_X_train = scaler_object.transform(X_train)
```

```python
scaled_X_test = scaler_object.transform(X_test)
```

```python
scaled_X_train
```

```python
    from keras.models import Sequential
    from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(scaled_X_train,y_train,epochs=150,verbose=2)
```

```python
model.predict_classes(scaled_X_test) # Does not predict one hot encoded result but only gives the class label
```

```python
model.predict(scaled_X_test) # Resulting in probabilities
```

```python
predictions = model.predict_classes(scaled_X_test)
```

```python
y_test.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
```

```python
confusion_matrix(y_test.argmax(axis=1),predictions)
```

```python
print(classification_report(y_test.argmax(axis=1),predictions))
```

```python
accuracy_score(y_test.argmax(axis=1),predictions)
```

```python
model.save('simplemodel.h5') # To save the weights of the model so that it can be reused again in some other model
```
